```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load cleaned reviews data
reviews_clean = pd.read_csv("/Users/thomassimmons/airbnb-boston-analysis/notebooks/data/cleaned/reviews_cleaned.csv")
reviews_clean["date"] = pd.to_datetime(reviews_clean["date"])
reviews_clean["year_month"] = reviews_clean["date"].dt.to_period("M").astype(str)

# Group by year-month to count reviews
monthly_reviews = reviews_clean.groupby("year_month").size().reset_index(name="review_count")
```

```python
# Plot
sns.set(style="whitegrid")
plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_reviews, x="year_month", y="review_count", marker="o")
plt.xticks(rotation=45, ha='right')
plt.title("Monthly Review Volume Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Reviews")
plt.tight_layout()
plt.savefig("/Users/thomassimmons/airbnb-boston-analysis/notebooks/figures/monthly_review_volume.png")
plt.show()
```
